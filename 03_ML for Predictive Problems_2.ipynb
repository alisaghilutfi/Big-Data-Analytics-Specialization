```python
import itertools

def readdata(k, fname="data/data.txt", report=False):
    C_k = []
    b = 0
    
    with open("groceries.csv", "rt", encoding='latin1') as f:
        lines = f.readlines()

    # lines = [l.split(',') for l in lines]
    lines = [l.replace('\n', '') for l in lines] #[i]  i in range(len(line(i)))]
    lines_ = []
    for i in lines:
        lines_ = [i.split(',') for i in lines]
        
    items =[]
    for l in lines_:
        for item in l:
            if item != "":
                # gather all items in one basket
                C_k.append(item)                 
            else: 
                # end of basket, report all itemsets
                for itemset in itertools.combinations(C_k, k):
                    yield frozenset(itemset)
                C_k = []
                
                # report progress
                # print every 1000th element to reduce clutter
                if report:
                    if b % 1000 == 0:  
                        print('processing bin ', b)
                    b += 1

    # last basket
    if len(C_k) > 0:
        for itemset in itertools.combinations(C_k, k):
            yield frozenset(itemset)    
```

```python
k= 2
import numpy as np
import os
import re
import binascii
from time import time
def get_shingles(f, k):
    """Get all shingles from requested file (hashes of these shingles)
    """
    items_ = ' '.join([line[:-1].strip() for line in f])
    items_ = re.sub(' +', ' ', items_)  # remove double spaces
    L = len(items_)
    shingles = set()  # we use a set to automatically eliminate duplicates
    for i in range(L-k+1):
        shingle = items_[i:i+k]
        crc = binascii.crc32(shingle.encode("utf-8")) #& 0xffffffff  # hash the shingle to a 32-bit integer
        shingles.add(crc)
    return shingles


```

```python
# set global parameters to process the whole dataset
### 100 minhashing function: bands*rows = 100
bands = 10
rows = 5
nsig = bands*rows  # number of elements in signature, or the number of different random hash functions

maxShingleID = 2**32-1  # record the maximum shingle ID that we assigned
nextPrime = 4294967311  # next prime number after maxShingleID

A = np.random.randint(0, nextPrime, size=(nsig,))
B = np.random.randint(0, nextPrime, size=(nsig,))
```

```python
# fast implementation of Minhash algorithm
# computes all random hash functions for a shingle at once, using vector operations
# also finds element-wise minimum of two vectors efficiently
def minhash_vectorized(shingles, A, B, nextPrime, maxShingleID, nsig):
    signature = np.ones((nsig,)) * (maxShingleID + 1)

    for ShingleID in shingles:
        hashCodes = ((A*ShingleID + B) % nextPrime) % maxShingleID
        np.minimum(signature, hashCodes, out=signature)

    return signature
```

```python
# get candidate pairs without Locality-Sensitive Hashing

def LSH(signatures, bands, rows, Ab, Bb, nextPrime, maxShingleID):
    """Locality Sensitive Hashing
    """
    numItems = (signatures).shape[1]
    print(numItems)
    signBands = np.array_split(signatures, bands, axis=0)
    candidates = set()
    for nb in range(bands):
        hashTable = {}
        for ni in range(numItems):
            item = signBands[nb][:,ni]
            hash = (np.dot(Ab[nb,:], item) + Bb[nb]) % nextPrime % maxShingleID
            if hash not in hashTable:
                hashTable[hash] = [ni]
            else:
                hashTable[hash].append(ni)
                
        for _,items in hashTable.items():
            if len(items) > 1:
                L = len(items)
                for i in range(L-1):  
                    for j in range(i+1, L):
                        cand = [items[i], items[j]]
                        np.sort(cand)
                        candidates.add(tuple(cand))
 
    return candidates
```

```python

signatures = []  # signatures for all files
t = time()
k = 2
for fname in lines:
    shingles = get_shingles(fname, k)
    signature = minhash_vectorized(shingles, A, B, nextPrime, maxShingleID, nsig)
    signatures.append(signature)
```

```python
# get candidate pairs without Locality-Sensitive Hashing

s = 0.9  # similarity threshold
Nfiles = len(signatures)
t = time()
candidates = []
for i in range(Nfiles):
    if time() - t > 18000:
        break
    else:
        for j in range(i+1, Nfiles):
            if time() - t < 18000:
                Jsim = np.mean(signatures[i] == signatures[j])  # average number of similar items in two vectors
                if Jsim >= s:
                    candidates.append((i,j))
            else:
                break
            
        
t2 = time() - t
```

```python
print("finding candidates took {} seconds".format(t2))
print("found {} candidates".format(len(candidates)))
print("candidate similar pairs of files are:")
for i,j in candidates[:20]:
    print(lines[i], lines[j])
```

```python
candidate_name = [[lines[i], lines[j]] for i,j in candidates]
```

```python
cluster = [[]]*(len(candidate_name))  # the maximum length of the cluster is equal to the size of similar items
n = 0
j = candidate_name[0][0] # start to compare the first file 
cluster[0] = [ candidate_name[0][0],  candidate_name[0][1]]  # the first cluster is the first two similar files, it can be included more files

for i in candidate_name[1:]:
#  the files are similar they are clustered in one cluster
    if (i[0] == j) and (i[1] not in cluster[n]) :
        if i[0] not in cluster[n]:
            cluster[n] = cluster[n] + [i[0]]
        cluster[n] = cluster[n] + [i[1]]          
    else:  # new cluster starts if the file is not similar to the previous files
        j = i[0]
        n = n + 1
cluster = [t for t in cluster if (t) != []]  # clusters which remained empty removed
print('number of clusters are',len(cluster))

```
